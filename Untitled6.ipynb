# Data Exploration

** Let's import the relevant libraries **

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

** And load in the data **

```python
df = pd.read_csv('data.csv')
```

** Next let's check the head, info, and description of the data set **

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

** We'll strip the full date down to just the year  **

```python
df['DOS'] = df['DOS'].str[:4]
```

```python
df['DOS'] = df['DOS'].astype(str).astype(int)
```

```python
dfx = df[df['DOS'] == 2017]
```

** Let's check out value_counts for 'LocationCode' and 'InsuranceFC' **

```python
df['LocationCode'].value_counts()
```

```python
df['InsuranceFC'].value_counts()
```

** We'll drop these non-numeric columns, as well as the 'id' column as it will simplify our work**

```python
df = df.drop(columns=['LocationCode', 'InsuranceFC','id'])
```

** We'll group the dataframe by the 'InsuranceCode' column **

```python
dfx = dfx.groupby(['InsuranceCode']).sum()
```

** and create a new column called 'PaymentsPerVisit' **

```python
dfx['PaymentsPerVisit'] = dfx['PaidToCharges'] / dfx['Visits']
```

** let's find out the top 15 most expensive insurance codes for the year 2017 **

```python
dfx.sort_values(by=['PaymentsPerVisit'], ascending=False).head(15)
```

```python
dfy = dfx.sort_values(by=['PaymentsPerVisit'], ascending=False)
```

** Let's create a graph displaying how much was paid per visit in 2017 across all insurance codes **

```python
## Paid Per Visit in 2017 by Insurance Code
ax = dfy[['PaymentsPerVisit']].plot(kind='barh', title ="Paid Per Visit", figsize=(30, 55), legend=True, fontsize=12)
ax.set_xlabel("Paid Per Visit", fontsize=12)
ax.set_ylabel("Insurance Code", fontsize=12)
```

** We'll drop some irrelevant values in order to simplify the conversion of InsuranceCode to a number **

```python
df = df.drop(df[df['InsuranceCode']=='V0322'].index)
df = df.drop(df[df['InsuranceCode']=='V7076'].index)
```

```python
df['InsuranceCode'] = df['InsuranceCode'].astype(str).astype(float)
```

```python
pd.options.display.max_rows = 100
```

```python
df[df['InsuranceCode']==10795]
```

** Some additional data exploration **

```python
yearly_summary = df.groupby(['DOS']).sum()
```

** Here are some statistics for each year **

```python
yearly_summary
```

```python
mean_charges = df.groupby(['DOS']).mean()
mean_charges.drop(columns=['Visits','NewPatients'])
```

** A graph showing the mean amount paid per visit for each year **

```python
ax = mean_charges[['PaidToCharges']].plot(kind='line', title ="Mean Amount Paid To Charges Per Visit Per Year", figsize=(20, 10), legend=True, fontsize=12, color='Green',grid=True)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Paid To Charges", fontsize=12)
```

** This graph shows the relationship between the number of visits and the amount paid to charges **

```python
df.plot.scatter(x='PaidToCharges', y='Visits', c='red', figsize=(20,10), grid=True, )
```

** And here's a visual representation for the number of visits per year **

```python
yearly_summary['Year'] = [2013,2014,2015,2016,2017,2018,2019]
```

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.style.use('fivethirtyeight')
ax = yearly_summary[['Visits']].plot(kind='bar', title ="Visits per Year", figsize=(20, 10), legend=True, fontsize=12, color='orange')
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Visits", fontsize=12)
```
